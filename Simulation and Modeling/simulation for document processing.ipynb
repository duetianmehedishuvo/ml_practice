```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame(columns=["jobs_no", "start_time", "work_time","finish_time","cumul_time","break_flag"])
worktimeList=[45,16,5,29,33,25,21]
artTime=0
worktime=worktimeList[0]
finishTime=worktime
cumulativeTime=worktime
breakFlag=0
numberOfJobs=57
if(cumulativeTime>60):
    breakFlag=1

df.loc[len(df)] = [numberOfJobs, startTime, worktime,finishTime,cumulativeTime,breakFlag]
for x in range(1, len(worktimeList)):
  numberOfJobs=numberOfJobs-1
  worktime=worktimeList[x]
  if breakFlag==1:
    startTime=finishTime+5
    cumulativeTime=worktime
  else:
    startTime=finishTime
    cumulativeTime=cumulativeTime+worktime

  finishTime=startTime+worktime
  
  if(cumulativeTime>60):
    breakFlag=1
  else:
    breakFlag=0

  df.loc[len(df)] = [numberOfJobs, startTime, worktime,finishTime,cumulativeTime,breakFlag]

  if(cumulativeTime>60):
    cumulativeTime=worktime
display(df)

# histogram of total_bills
plt.hist(df.cumul_time)
 
plt.title("Histogram")
 
# Adding the legends
plt.show()

sns.histplot(x='cumul_time', data=df, kde=True)
 

```
